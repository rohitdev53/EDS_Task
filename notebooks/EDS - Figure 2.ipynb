```python
owid_data_path='../data/raw/owid-covid-data.csv'
pd_raw_owid=pd.read_csv(owid_data_path)
```

```python
def vac(country_name):
    data = pd_raw_owid[pd_raw_owid['location']==country_name]
    x=data['date']
    pop=data['population']
    tot_vac_rate=data['total_vaccinations'].diff()
    return tot_vac_rate/pop*100, data['date']
```

```python
def add_graph(country_name):
    y,x = vac(country_name)
    fig.add_trace(go.Scatter(x=x,
                            y=y,
                            mode='markers+lines',
                            opacity=0.4,
                            line_width=1,
                            marker_size=1, 
                            name=country_name))
    
    fig.add_trace(go.Scatter(x=x,
                        y=y.rolling(7).mean(),
                        mode='markers+lines',
                        opacity=1,
                        line_width=2,
                        marker_size=1, 
                        name=country_name+'_mean'))
```

```python
fig = go.Figure()

fig = make_subplots()
add_graph('Germany')
add_graph('India')
add_graph('Poland')

## defines the overall layout properties
fig.update_layout(
    width=1000,
    height=900,
    xaxis_title="Time Period (in days)",
    yaxis_title="Absolute Covid cases/population size",
)


fig.update_layout(xaxis_rangeslider_visible=True)
fig.show()
```
